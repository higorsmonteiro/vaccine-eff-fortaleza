## Plot risks and effectiveness

```python
def plot_risks(tb_final):
    fig, (ax1, ax2) = plt.subplots(1,2, figsize=(12,4.5))

    sns.lineplot(x='t_f(caso)', y='1 - KM(caso)', data=tb_final, color="tab:blue", lw=2, ax=ax1, label="vaccinated")
    sns.lineplot(x='t_f(controle)', y='1 - KM(controle)', data=tb_final, color="tab:red", lw=2, ax=ax1, label="unvaccinated")
    sns.lineplot(x='t_f(caso)', y='1 - Risk Ratio', data=tb_final, color="tab:green", lw=2, ax=ax2)

    ax1.legend(prop={'size':13})
    for axis in [ax1, ax2]:
        axis.grid(alpha=0.3)
        axis.set_xlabel("Time to event", fontsize=12)
    
    return fig, (ax1,ax2)

def plot_risks_compare(tb_final1, tb_final2, str_1="1", str_2="2"):
    fig, (ax1, ax2) = plt.subplots(1,2, figsize=(12,4.5))

    sns.lineplot(x='t_f(caso)', y='1 - KM(caso)', data=tb_final1, color="tab:blue", lw=2, ax=ax1, ls="--", label=f"vaccinated({str_1})")
    sns.lineplot(x='t_f(controle)', y='1 - KM(controle)', data=tb_final1, color="tab:red", lw=2, ax=ax1, ls="--", label=f"unvaccinated({str_1})")
    sns.lineplot(x='t_f(caso)', y='1 - Risk Ratio', data=tb_final1, color="tab:green", lw=2, ax=ax2, ls="--", label=f"({str_1})")

    sns.lineplot(x='t_f(caso)', y='1 - KM(caso)', data=tb_final2, color="tab:blue", lw=2, ax=ax1, label=f"vaccinated({str_2})")
    sns.lineplot(x='t_f(controle)', y='1 - KM(controle)', data=tb_final2, color="tab:red", lw=2, ax=ax1, label=f"unvaccinated({str_2})")
    sns.lineplot(x='t_f(caso)', y='1 - Risk Ratio', data=tb_final2, color="tab:orange", lw=2, ax=ax2, label=f"({str_2})")

    ax1.set_ylabel("1 - KM", fontsize=12)
    ax2.set_ylabel("1 - Risk Ratio", fontsize=12)
    ax2.set_ylim([0.6, 1.01])
    for axis in [ax1, ax2]:
        axis.grid(alpha=0.3)
        axis.set_xlabel("Time to event", fontsize=12)
        axis.legend(prop={'size':13})
        
    return fig, (ax1,ax2)
```

```python
vacinados_df = pd.read_csv(os.path.join("..", "..", "..", "data", "vacinado_obito", "vacinados_d1d2_update_Sep2_2021.csv"))
fe_total, fe_d1, fe_d1d2 = resume_vacinados(vacinados_df)

dff = pd.read_excel(os.path.join("..", "..", "..", "data", "comparativo_vacinado_pop_faixa_1.xlsx"))
dff = dff.rename({"Unnamed: 0": "FAIXA"}, axis=1)
dff["interval"] = dff["FAIXA"].apply(lambda x: x.replace("ANOS","").replace("A", ",").replace(",NO", "").replace("M,IS", "+"))
dff["FAIXA MIN"] = dff["interval"].apply(lambda x: x.split(",")[0] if len(x.split(","))==2 else ("0" if "<" in x else ("100")))
dff["FAIXA MAX"] = dff["interval"].apply(lambda x: x.split(",")[1] if len(x.split(","))==2 else ("1" if "<" in x else ("120")))
dff["FAIXA MAX"] = dff["FAIXA MAX"].astype(int)
dff["FAIXA MIN"] = dff["FAIXA MIN"].astype(int)
dff = dff.drop(["interval", "Qnt %", "faixa vacinados"], axis=1)

dff1 = dff.merge(fe_total, left_on="FAIXA", right_on="FAIXA TOTAL", how="left").drop("FAIXA TOTAL", axis=1)
dff1 = dff1.merge(fe_d1, left_on="FAIXA", right_on="FAIXA ONLY D1", how="left").drop("FAIXA ONLY D1", axis=1)
dff1 = dff1.merge(fe_d1d2, left_on="FAIXA", right_on="FAIXA D1+D2", how="left").drop("FAIXA D1+D2", axis=1)
```

```python
def resume_vacinados(vacinados_df):
    '''
    
    '''
    vacinados_df["VACINA STATUS"] = vacinados_df[["data D1", "data D2"]].apply(lambda x: "ONLY D1" if pd.isna(x["data D2"]) and not pd.isna(x["data D1"]) else ("D1+D2" if not pd.isna(x["data D2"]) and not pd.isna(x["data D1"]) else "SEM"), axis=1)
    condition_total = (vacinados_df["VACINA STATUS"]=="ONLY D1") | (vacinados_df["VACINA STATUS"]=="D1+D2")
    condition_d1 = vacinados_df["VACINA STATUS"]=="ONLY D1"
    condition_d1d2 = vacinados_df["VACINA STATUS"]=="D1+D2"
    
    faixas_total = vacinados_df[condition_total]["faixa etaria"].value_counts().reset_index().rename({"index": "faixa vacinado", "faixa etaria": "TOTAL"}, axis=1)
    faixas_d1 = vacinados_df[condition_d1]["faixa etaria"].value_counts().reset_index().rename({"index": "faixa vacinado", "faixa etaria": "ONLY D1"}, axis=1)
    faixas_d1d2 = vacinados_df[condition_d1d2]["faixa etaria"].value_counts().reset_index().rename({"index": "faixa vacinado", "faixa etaria": "D1+D2"}, axis=1)
    faixas_total["FAIXA TOTAL"] = faixas_total["faixa vacinado"].apply(lambda x: x.replace(" ", "")).replace(to_replace="100E+ANOS", value="100MAISANOS")
    faixas_d1["FAIXA ONLY D1"] = faixas_d1["faixa vacinado"].apply(lambda x: x.replace(" ", "")).replace(to_replace="100E+ANOS", value="100MAISANOS")
    faixas_d1d2["FAIXA D1+D2"] = faixas_d1d2["faixa vacinado"].apply(lambda x: x.replace(" ", "")).replace(to_replace="100E+ANOS", value="100MAISANOS")
    faixas_total = faixas_total.drop("faixa vacinado", axis=1)
    faixas_d1 = faixas_d1.drop("faixa vacinado", axis=1)
    faixas_d1d2 = faixas_d1d2.drop("faixa vacinado", axis=1)
    return faixas_total, faixas_d1, faixas_d1d2
    
def new_table(dff, intervals):
    '''
    
    '''
    new_table = {"FAIXA": [], "ESTIMATIVA 2021": [], "VACINADOS TOTAL": [],
                "VACINADOS SOMENTE D1": [], "VACINADOS D1+D2": []}
    for interval in intervals:
        str_anos = f"{interval[0]} A {interval[1]} ANOS"
        if interval[1]>100:
            str_anos = f"{interval[0]}+ ANOS"
        condition = (dff["FAIXA MIN"]>=interval[0]) & (dff["FAIXA MAX"]<=interval[1]) 
        sub = dff[condition]
        new_table["FAIXA"].append(str_anos)
        new_table["ESTIMATIVA 2021"].append(sub["ESTIMATIVA 2021"].sum())
        new_table["VACINADOS TOTAL"].append(sub["TOTAL"].sum())
        new_table["VACINADOS SOMENTE D1"].append(sub["ONLY D1"].sum())
        new_table["VACINADOS D1+D2"].append(sub["D1+D2"].sum())
    return pd.DataFrame(new_table)
```

```python
df2 = new_table(dff1, [[0,9], [10,19], [20,59], [60,120]])

df2["VACINADOS TOTAL %"] = (df2["VACINADOS TOTAL"]/df2["ESTIMATIVA 2021"])*100
df2["VACINADOS SOMENTE D1 %"] = (df2["VACINADOS SOMENTE D1"]/df2["ESTIMATIVA 2021"])*100
df2["VACINADOS D1+D2 %"] = (df2["VACINADOS D1+D2"]/df2["ESTIMATIVA 2021"])*100
```
