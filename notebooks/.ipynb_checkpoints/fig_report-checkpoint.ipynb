## **Lib**

```python
import os
import sys
sys.path.append("..")

import numpy as np
import pandas as pd
import datetime as dt
import lib.utils as utils
import lib.fig_utils as fig_utils
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from src.MatchedCount import MatchedCount
from lifelines import KaplanMeierFitter

import warnings
warnings.filterwarnings('ignore')

init_cohort = dt.datetime(2021, 1, 21)
final_cohort = dt.datetime(2021, 8, 31)
```

## **Define data and metadata**

```python
path_to = os.path.join("..", "output")
data_name = {
    "GENERAL POPULATION": "ELIGIBILIDADE_CADASTRADOS_VACINAS_COORTE_FORTALEZA_1634587254.csv",
    "MATCHED CORONAVAC": "pareados_corona.csv", 
    "MATCHED ASTRAZENECA": "pareados_astra.csv", 
    "MATCHED PFIZER": "pareados_pfizer.csv", 
    "SURVIVAL DATA CORONA": "sobrevida_coronavac.csv",
    "SURVIVAL DATA ASTRA": "sobrevida_astrazeneca.csv",
    "POPULATION FOR CORONA": "control_reservoir_corona.csv", 
    "POPULATION FOR ASTRA": "control_reservoir_astra.csv" 
}

general_pop = pd.read_csv(os.path.join(path_to, data_name["GENERAL POPULATION"]), index_col=0)
general_pop["CPF"] = general_pop["CPF"].astype(str, errors="ignore")
general_pop["CPF"] = general_pop["CPF"].apply(lambda x: x.split(".")[0] if not pd.isna(x) else np.nan)
```

**Load data - CoronaVac**

```python
matched_corona = pd.read_csv(os.path.join(path_to, data_name["MATCHED CORONAVAC"]), index_col=0)
survival_corona = pd.read_csv(os.path.join(path_to, data_name["SURVIVAL DATA CORONA"]))
pop_corona = pd.read_csv(os.path.join(path_to, data_name["POPULATION FOR CORONA"]), index_col=0)
```

**Load data - Astrazeneca**

```python
matched_astra = pd.read_csv(os.path.join(path_to, data_name["MATCHED ASTRAZENECA"]), index_col=0)
survival_astra = pd.read_csv(os.path.join(path_to, data_name["SURVIVAL DATA ASTRA"]))
pop_astra = pd.read_csv(os.path.join(path_to, data_name["POPULATION FOR ASTRA"]), index_col=0)
```

**Columns for each data**

```python
df_columns = pd.concat([pd.Series(matched_corona.columns), pd.Series(survival_astra.columns), 
           pd.Series(pop_astra.columns), pd.Series(general_pop.columns)], axis=1)
df_columns.columns = ["DADO PAREADOS", "DADO SOBREVIDA", "DADO POPULACAO VACINA", "DADO POPULACAO GERAL"]
df_columns
```

**Survival analysis**

```python
obj_deaths = MatchedCount(os.path.join("..", "output", "sobrevida_coronavac.csv"), "OBITO")
obj_hospt = MatchedCount(os.path.join("..", "output", "sobrevida_coronavac.csv"), "HOSPITALIZACAO")
obj_tests = MatchedCount(os.path.join("..", "output", "sobrevida_coronavac.csv"), "TESTE")

KM_CASO_DEATH, KM_CONTROLE_DEATH = obj_deaths.fit_km_period(init_period=0, negative_intervals=False)
KM_CASO_HOSPT, KM_CONTROLE_HOSPT = obj_hospt.fit_km_period(init_period=0, negative_intervals=False)
KM_CASO_TESTS, KM_CONTROLE_TESTS = obj_tests.fit_km_period(init_period=0, negative_intervals=False)

KM_CASO_DEATH = KM_CASO_DEATH.iloc[:100, :]
KM_CASO_HOSPT = KM_CASO_HOSPT.iloc[:100, :]
KM_CASO_TESTS = KM_CASO_TESTS.iloc[:100, :]
KM_CONTROLE_DEATH = KM_CONTROLE_DEATH.iloc[:100, :]
KM_CONTROLE_HOSPT = KM_CONTROLE_HOSPT.iloc[:100, :]
KM_CONTROLE_TESTS = KM_CONTROLE_TESTS.iloc[:100, :]
```

```python
fig, AX = utils.KM_plotting(KM_CASO_DEATH, KM_CONTROLE_DEATH, KM_CASO_HOSPT, KM_CONTROLE_HOSPT)
fig.suptitle("CORONAVAC - D1 and D1+D2", fontsize=14, y=0.92)
#fig.savefig(os.path.join("..", "output", "report", "KM_d1_d1d2.pdf"), dpi=150)
```

**1 - Hazard Ratio**

```python
fig, ax1 = plt.subplots(1,figsize=(8,6))

day_death = KM_CASO_DEATH["day"].tolist()
day_hospt = KM_CASO_HOSPT["day"].tolist()
ve_death = 1 - (KM_CASO_DEATH["KM_estimate"]/KM_CONTROLE_DEATH["KM_estimate"])
ve_hospt = 1 - (KM_CASO_HOSPT["KM_estimate"]/KM_CONTROLE_HOSPT["KM_estimate"])

ax1.plot(day_death, ve_death, label="Death by Covid-19", lw=2)
ax1.plot(day_hospt, ve_hospt.iloc[:-1], label="Hospitalization due to Covid-19", lw=2)

ax1.set_xlabel("days", fontsize=13)
ax1.set_ylabel("1 - Hazard Ratio", fontsize=13)
ax1.set_xlim([7,80])
ax1.set_ylim([0.0,1.0])
ax1.set_xticks(np.arange(7, 90+1, 7))
ax1.legend(loc=3, prop={"size":13})
ax1.grid(alpha=0.2)
fig.suptitle("CORONAVAC - D1 and D1+D2")
df = pd.concat([ve_hospt, ve_death], axis=1)
df
#fig.savefig(os.path.join("..", "output", "report", "daily_1_HR_d1_d1d2.pdf"), dpi=150)
```
