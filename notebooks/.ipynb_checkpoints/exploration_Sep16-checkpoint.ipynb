## Exploração dos dados de testes GAL

```python
# Carregar testes
testes_fname = os.path.join(data_paths["TESTES COVID-19"], data_names["TESTES COVID-19"])
colnames = ["Index tabela original", "Municipio do Solicitante", "Estado do Solicitante",
           "Data da Solicitação", "Data Notificação Sinan", "Data Notificação Gal",
           "Data de Nascimento", "Idade", "Bairro", "Data da Coleta", "1º Campo Resultado"]
testes_df = dutils.open_testes_gal(testes_fname, colnames=colnames, delimiter=",")
testes_df.head()
```

```python
# Transformação e filtros dos dados
testes_df = testes_df[testes_df["Municipio do Solicitante"]=="FORTALEZA"]
df = testes_df

def f(x):
    res_str = ["Resultado: Não Detectável", "Resultado: Detectável"]
    if pd.isna(x): return "OUTROS"
    if x==res_str[0]:
        return "NEGATIVO"
    elif x==res_str[1]:
        return "POSITIVO"
    else:
        return "OUTROS"

df["RESULTADO FINAL"] = df["1º Campo Resultado"].apply(lambda x: f(x))
```
