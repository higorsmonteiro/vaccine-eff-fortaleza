```python
fpath = os.path.join("..", "output", "data", "LINKAGE")
fpathdata = os.path.join("..", "..", "..", "data", "PARQUET_TRANSFORMED")

# --> Linkage data
link_ja_vac = pd.read_parquet(os.path.join(fpath, "VACINEJA_VACINADOS.parquet"))
link_ja_int = pd.read_parquet(os.path.join(fpath, "VACINEJA_INTEGRASUS.parquet"))
link_ja_covid = pd.read_parquet(os.path.join(fpath, "VACINEJA_OBITO_COVID.parquet"))
link_ja_cart = pd.read_parquet(os.path.join(fpath, "VACINEJA_OBITO_CARTORIO.parquet"))

# --> Databases
vacineja_df = pd.read_parquet(os.path.join(fpathdata, "VACINEJA.parquet"))
vacinados_df = pd.read_parquet(os.path.join(fpathdata, "VACINADOS.parquet"))
integra_df = pd.read_parquet(os.path.join(fpathdata, "INTEGRASUS.parquet"))
covid_df = pd.read_parquet(os.path.join(fpathdata, "OBITO_COVID.parquet"))
cart_df = pd.read_parquet(os.path.join(fpathdata, "OBITO_CARTORIO.parquet"))
```

```python
found_vacinados = link_ja_vac[pd.notna(link_ja_vac["cpf(VACINEJA)"])]
found_tests = link_ja_int[pd.notna(link_ja_int["cpf(VACINEJA)"])]
found_covid = link_ja_covid[pd.notna(link_ja_covid["cpf(VACINEJA)"])]
found_cart = link_ja_cart[pd.notna(link_ja_cart["cpf(VACINEJA)"])]

to_vacinados = defaultdict(lambda: np.nan, zip(found_vacinados["cpf(VACINEJA)"], found_vacinados["cpf(VACINADOS)"]))
to_tests = defaultdict(lambda: np.nan, zip(found_tests["cpf(VACINEJA)"], found_tests["id"]))
to_covid = defaultdict(lambda: np.nan, zip(found_covid["cpf(VACINEJA)"], found_covid["numerodo"]))
to_cart = defaultdict(lambda: np.nan, zip(found_cart["cpf(VACINEJA)"], found_cart["cpf(CARTORIOS)"]))
```

```python
vacineja_df["cpf LINKAGE VACINADOS"] = vacineja_df["cpf"].apply(lambda x: to_vacinados[x])
vacineja_df["id LINKAGE INTEGRASUS"] = vacineja_df["cpf"].apply(lambda x: to_tests[x])
vacineja_df["numerodo LINKAGE OBITO COVID"] = vacineja_df["cpf"].apply(lambda x: to_covid[x])
vacineja_df["cpf LINKAGE CARTORIOS"] = vacineja_df["cpf"].apply(lambda x: to_cart[x])
```

```python
integra_df = integra_df.set_index("id")
```

```python
def select_indexes(integra_df, indexes):
    '''
    
    '''
    selected = integra_df.loc[indexes]
    #print(selected["INFO COORTE SINTOMAS"].values, selected["INFO COORTE SOLICITACAO"].values, selected["INFO COORTE COLETA"].values)
    if np.isin(["SIM"], selected["INFO COORTE SINTOMAS"].values)[0]:
        return "SIM"
    if np.isin(["SIM"], selected["INFO COORTE SOLICITACAO"].values)[0]:
        return "SIM"
    if np.isin(["SIM"], selected["INFO COORTE COLETA"].values)[0]:
        return "SIM"
    return "NAO"
```

```python
%timeit
vacineja_df["TESTE POSITIVO ANTES COORTE"] = vacineja_df["id LINKAGE INTEGRASUS"].apply(lambda x: select_indexes(integra_df, x) if type(x)!=float else "NAO")
```
