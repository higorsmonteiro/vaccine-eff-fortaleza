```python
def cohort_vaccine(x, init_cohort, final_cohort):
    '''
         Vaccination status during cohort.
    '''
    d1 = x['DATA D1']
    d2 = x['DATA D2']
    d3 = x['DATA D3']
    
    str_vac = ''
    if pd.notna(d1) and (d1>=init_cohort and d1<=final_cohort):
        str_vac+='(D1)'
    if pd.notna(d2) and (d2>=init_cohort and d2<=final_cohort):
        str_vac+='(D2)'
    if pd.notna(d3) and (d3>=init_cohort and d3<=final_cohort):
        str_vac+='(D3)'
    return str_vac
    
```

```python
df["STATUS VACINACAO NA COORTE"] = df[["DATA D1", "DATA D2", "DATA D3"]].apply(lambda x: cohort_vaccine(x, dt.datetime(2021, 1, 21), dt.datetime(2021, 8, 31)), axis=1)
```
