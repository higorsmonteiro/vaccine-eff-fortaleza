# LIB

```python
import os
import pandas as pd
import numpy as np
import datetime as dt
from collections import defaultdict

import matplotlib.pyplot as plt
%matplotlib inline
```

# Load data

```python
tests = ("CORONAPOP", "TODAPOP")
vaccine = "CORONAVAC"
data_folder = os.path.join("..", "..", "..", "data")
output_folder = os.path.join("..", "output", "data")
pareamento_folder = os.path.join("..", "output", "PAREAMENTO")
survival_folder = os.path.join("..", "output", "PAREAMENTO", vaccine, "SURVIVAL")

cohort_period = (dt.datetime(2021, 1, 21), dt.datetime(2021, 8, 31))
cohort_str = ("21JAN2021", "31AUG2021")
fname_schema = f"SCHEMA_{cohort_str[0]}_{cohort_str[1]}.parquet"

fschema = pd.read_parquet(os.path.join(output_folder, fname_schema))
```

```python
seed = 1
t_min = ["0","14"]
event = ["OBITO", "HOSPITAL", "UTI"]
coronapop_ve = {
        "OBITO": {"0": None, "14": None},
        "HOSPITAL": {"0": None, "14": None},
        "UTI": {"0": None, "14": None},
}
todapop_ve = {
        "OBITO": {"0": None, "14": None},
        "HOSPITAL": {"0": None, "14": None},
        "UTI": {"0": None, "14": None},
}
km_coronapop = {
    "OBITO": None,
    "HOSPITAL": None,
    "UTI": None,
}
km_todapop = {
    "OBITO": None,
    "HOSPITAL": None,
    "UTI": None,
}

for ev in event:
    km_coronapop[ev] = pd.read_excel(os.path.join(survival_folder, f"SEED{seed}CORONAPOP", f"KM_events_0_{ev}.xlsx"), sheet_name=None)
    km_todapop[ev] = pd.read_excel(os.path.join(survival_folder, f"SEED{seed}TODAPOP", f"KM_events_0_{ev}.xlsx"), sheet_name=None)
    for t in t_min:
        coronapop_ve[ev][t] = pd.read_excel(os.path.join(survival_folder, f"SEED{seed}CORONAPOP", f"VE_{t}_{ev}.xlsx"), sheet_name=None)
        todapop_ve[ev][t] = pd.read_excel(os.path.join(survival_folder, f"SEED{seed}TODAPOP", f"VE_{t}_{ev}.xlsx"), sheet_name=None)
```

# Plot VE

```python
evento = "OBITO"
curdict = coronapop_ve
df1 = curdict[evento]["14"]["D2"]
fig, ax = plt.subplots(1)

ax.plot(df1["t"], df1["VE"], lw=2, color="orange")
ax.fill_between(df1["t"], df1["VE_lower_0.95"], df1["VE_upper_0.95"], alpha=0.2, color="orange")

ax.set_xlim([14,150])
ax.set_ylim([.0,1])
ax.set_yticks(np.arange(0, 1.1, 0.1))

ax.grid(alpha=0.3)
```
